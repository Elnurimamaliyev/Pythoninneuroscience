## Exercise 2
Write a function that computes the value of the following mathematical function: $f(x,y) = x^2 + y$

Plot the results for
100 x-values covering the range of [-2.0, 2.0]
and three different, but constant, values of y = [-1, 0, 1]. 
This should produce three curves, each with 100 data points.

Annotate the plot with proper labels and a legend that indicates the vale of y.

Roughly like so:

![](2.03_functions_exercises.png)

```python
# your solution here
import matplotlib.pyplot as plt
import numpy as np


xlist = np.linspace(-2.0, 2.0, 100)
# ylist= np.linspace(-1, 1, 3)

ylist= [-1,0,1]

def f(x,y):
    equation = (x**2)+y
    return equation

for y in ylist:
    results = [f(x, y) for x in xlist]
    plt.plot(xlist, results, label=f"y={int(y)}")

plt.xlabel('x')
plt.ylabel('f(x, y)')
plt.title(r'$f(x, y) = x^2 + y$')
plt.legend()
plt.show()

```

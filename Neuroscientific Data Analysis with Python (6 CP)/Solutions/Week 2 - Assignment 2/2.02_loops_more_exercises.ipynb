## Exercise 4 - `zip`
You are given to lists - names with associated phone numbers.

Print each name and the associated phone number, like so:
```
Alice 432246234
Yolanda 190998010
Tom 984703
Francoise 123156
Alan 9630
```

Use a for loop and `zip` to loop over both lists in parallel. Remember, `zip` takes two (or more) lists as input and returns the items for both lists:
```python
for item_list1, item_list2 in zip(list1, list2)
    do something
```


````python
names = ['Alice', 'Yolanda', 'Tom', 'Francoise', 'Alan']
phone_numbers = [432246234, 190998010, 984703, 123156, 9630]

# your solution here
for name, phone_number in zip(names, phone_numbers):
    print(name, phone_number)
````

## Exercise 5 - Make a digital phone book
Make a phone book using a dictionary, so we can easily look up phone numbers using the name as the key.
We are given two parallel lists of names and phone numbers. Generate a phone book dictionary that has each name as a key and the associated phone number as a value. Use `zip` to loop over the parallel lists.

````python
names = ['Alice', 'Yolanda', 'Tom', 'Francoise', 'Alan']
phone_numbers = [432246234, 190998010, 984703, 123156, 9630]

# your solution here
phone_dictionary = {}

for name, phone_number in zip(names, phone_numbers):
    phone_dictionary[name] = phone_number

print(phone_dictionary)

````

## Exercise 7 - filtering a list

We'd like to produce a plot that highlights when neg. and pos. anomaly values occurred.

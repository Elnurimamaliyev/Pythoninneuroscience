## Range
The `range` function generates a sequence of numbers:

```python
for index in range(4):
    print(index)
```

Why is this useful? For instance to use as indices in a for loop.

If you have a list, you can generate a range of the same length as your list for indexing:

```python
data = [10234,234,523,52,34, 234]

for index in range(len(data)):
    print(index, data[index])
```

One can also use `range` to create a long list of numbers and convert it's units to use for plotting.

Say we have measurements, taken along a line at intervals of 200 kilometers. We can use `range` to generate the x-values for our plot:

```python
temperature = [14, 32, 23, 18]

positions_km = []
for index in range(len(temperature)):
    positions_km.append(index * 200)

import matplotlib.pyplot as plt
plt.plot(positions_km, temperature, 'o-')
plt.xlabel('Position [km]')
plt.ylabel('Temperature [C]')
plt.show()
```

## Plotting
The order of the commands matters - where you put `plt.figure`, `plt.subplot`, `plt.plot()` and `plt.show()` determines what is plotted where:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))  # first generate a figure with a specific size

plt.subplot(1, 2, 1)  # a call to subplot after that will add the subplot to that figure
plt.plot([1, 0], 'r')  # plot will now plot into the subplot

plt.subplot(1, 2, 2)  # make a second subplot in the same figure
plt.plot([1, 2], 'g')  # plot into the second subplot

plt.show()  # display the figure. This will show the figure on screen but "delete" the figure in python. Any calls to plt.subplot or plt.plot after plt.show() will use a new figure!!
```
